# Ejemplo de la aplicaciГіn de Deep Learning para un problema de clasificaciГіn con Keras
Keras es una herramienta fГЎcil de utilizar y forma parte del paquete TensorFlow. Se utiliza para desarrollar y evaluar modelos de aprendizaje profundo.

TensorFlow es una poderosa librerГ­a que permite la definiciГіn y entrenamiento de modelos de redes neuronales en pocas lГ­neas de cГіdigo.

En este ejemplo, vamos a utilizar Keras para entrenar y evaluar un modelo para la clasificaciГіn de pacientes con diabetes.

**Los pasos que vamos a trabajar son los siguientes:**
1. Cargar los datos y prepararlos para utilizarlos en los modelos
1. Definir el modelo de Keras
1. Compilar el modelo
1. Ajustar el modelo con los datos
1. Evaluar la precisiГіn del modelo
1. Realizar predicciones

Primero, debemos de importar las librerГ­as necesarias, lo cual realizamos en el siguiente segmento de cГіdigo.

```python
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
```

# 1. Carga de datos
Para cargar los datos, el dataset debe de ser incluГ­do en el Notebook, para ello nos aseguramos que se encuentre en el directorio de fuentes con el cГіdigo inicial generado por Kaggle.

La descripciГіn de las columnas es la siguiente:
1. NГєmero de veces que quedo embarazada
1. ConcentraciГіn de glucosa luego de 2 horas de ingestiГіn glucosa
1. PresiГіn arterial diastГіlica (mm Hg)
1. Grosor del pliegue cutГЎneo del tricep (mm)
1. 2-hour despues del serum insulina (mu U/ml)
1. Indice de masa muscular (weight in kg/(height in m)^2)
1. FunciГіn del pedigrГ­ de la diabetes
1. Edad
1. Si tiene diabetes o no (1 o 0)

Posteriormente separamos el vector en las variables X y Y.

```python
# utilizamos la funciГіn loadtxt para cargar la informaciГіn del CSV
dataset = loadtxt('/content/dataset_diabetes.csv', delimiter=',')

#tomamos todos los registros de la columna 0 a la columna 7 (variables independientes)
X = dataset[:, 0:8]
#tomamos todos los registros de la columna 8 (variable dependiente)
y = dataset[:,8]
```

```python
y.shape
```

```python
X_train, X_test = X[:615,:], X[615:,:]
y_train, y_test = y[:615], y[615:]
```

```python
X_train.shape
```

# 2. Definir el modelo de Keras

Para la definiciГіn del modelo, vamos a utilizar una secuencia de capaz. Para ello vamos a utilizar un modelo secuencial.

Regularmente, se definen ese tipo de modelos para problemas de clasificaciГіn y se agrega una a una las capaz que consideremos necesarias para alcanzar los objetivos que deseamos.

En cuanto a la cantidad de capaz, no existe un parГЎmetro que podamos calcular o definir a ciencia cierta cual es el nГєmero Гіptimo. Regularmente se  realizan procesos de prueba y error (como se viГі en el ciclo de la presentaciГіn de la clase) para determinar un nГєmero de capaz con el que estemos satisfechos.

Para este ejemplo, vamos a utilizar una estructura completamente conectada en la red neuronal. Para ello se utiliza la clase Dense de Keras.

En cada capa es posible asignar un nГєmero de nodos por capa y la funciГіn de activaciГіn.

```python
# DefiniciГіn del modelo
model = Sequential()
# Agregando la primer capa
model.add(Dense(128, activation='relu'))
# Agregando la primer capa
model.add(Dense(64, activation='relu'))
# Agregando la segunda capa
model.add(Dense(32, activation='relu'))
# Agregando la tercer capa
model.add(Dense(16, activation='relu'))
# Agregando la Гєltima capa, la cual predice y_hat con la funciГіn sigmoid como activaciГіn
model.add(Dense(1, activation='sigmoid'))
```

## 3. Compilar el modelo
Luego de definir el modelo, se debe de compilar para poder utilizarlo. El proceso de compilaciГіn del modelo internamente crea un backend que elige cual es la mejor forma de representar la red para entrenamiento y se configura de tal forma que se pueda utilizar para entrenamiento y realizar predicciones.

Al compilar se deben de especificar ciertos parГЎmetros como la Loss Function, el optimizador y la mГ©tricas que serГЎn utilizadas para la evaluaciГіn del rendimiento (lo cual recolecta y reporta la precisiГіn de la clasificaciГіn).

```python
# Compilando el modelo, con la funciГіn de perdida "binary_crossentropy" por ser un problema de clasificaciГіn binario.
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## 4. Entrenar o ajustar el modelo
Luego de definir y compilar el modelo, podemos continuar con el entrenamiento del modelo. Para entrenar el modelo utilizamos la funciГіn fit().

El entrenamiento se realizar por "Г©pocas" y por "lotes". Las epocas se refiere a una pasada por todos los datos de entrenamiento y los lotes son la cantidad de registros dentro de una Г©poca antes de considerar actualizar los pesos.

```python
# Ajustando el modelo, con sin corridas completas de todos los datos y actualizando los pesos cada 10 registros
model.fit(X_train, y_train, epochs=100, batch_size=50, verbose=1)
```

# 5. Evaluando el modelo
Para evaluar el modelo que hemos entrenado, podemos evaluar el rendimiento en el mismo dataset de entrenamiento o en un dataset para pruebas. Para evaluar el rendimiento utilizamos la funciГіn evaluate.

La funciГіn evaluate retorna la perdida y la precisiГіn.

```python
# Evaluando el rendimiento del modelo
loss, accuracy = model.evaluate(X_train, y_train)
print('PrecisiГіn: %.2f' %(accuracy*100))
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print('PrecisiГіn: %.2f' %(accuracy*100))
```

# 6. Predicciones
Para realizar predicciones utilizamos el mГ©todo predict(). Como se estГЎ utilizando una funciГіn sigmoid en la Гєltima capa de la red neuronal, el valor de y_hat va a estar entre 0 y 1. Para obtener el valor de y_hat igual a 0 o 1, debemos de transformarlo.

```python
# PredicciГіn con transformaciГіn de las salidas:
predictions = model.predict(X)
# transformaciГіn de las salidas  utilizando la tГ©cnica de redondeo
#rounded = [round(x[0]) for x in predictions]

predictions[0:10]
```

```python
# De forma alternativa, podemos realizar la transformaciГіn directamente en la predicciГіn:
predictions = (model.predict(X) > 0.40).astype(int)
predictions[0:10]
```

```python
# A continuaciГіn se muestran los primero diez registros y sus predicciones:
for i in range(10):
    print('%s => %d (esperado %d)' % (X[i].tolist(), predictions[i][0], y[i]))
```
